```python
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import tensorflow as tf

# Load the activity recognition model (pre-trained or custom)
model = tf.keras.applications.MobileNetV2(weights='imagenet', include_top=True)

def load_video(video_path):
    cap = cv2.VideoCapture(video_path)
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        yield frame
    cap.release()

def analyze_frame(frame):
    # Resize and prepare frame for the model
    resized_frame = cv2.resize(frame, (224, 224))
    frame_array = tf.keras.applications.mobilenet_v2.preprocess_input(resized_frame)
    frame_array = np.expand_dims(frame_array, axis=0)

    predictions = model.predict(frame_array)
    decoded_predictions = tf.keras.applications.mobilenet_v2.decode_predictions(predictions, top=1)[0]
    return decoded_predictions[0][1]  # Return the top prediction

def generate_caption(activity):
    return f"Activity Detected: {activity}"

def embed_caption(frame, caption):
    img = Image.fromarray(frame)
    draw = ImageDraw.Draw(img)
    font = ImageFont.truetype("arial.ttf", 20)
    draw.text((10, 10), caption, font=font, fill="white")
    return np.array(img)

def save_video(frames, output_path):
    height, width, layers = frames[0].shape
    video = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'MP4V'), 24, (width, height))
    for frame in frames:
        video.write(frame)
    video.release()

def process_video(video_path):
    output_path = video_path.replace('.mp4', '_captioned.mp4')
    processed_frames = []
    for frame in load_video(video_path):
        activity = analyze_frame(frame)
        caption = generate_caption(activity)
        captioned_frame = embed_caption(frame, caption)
        processed_frames.append(captioned_frame)
    
    save_video(processed_frames, output_path)
    return output_path

# Example
video_path = '7413734-hd_1920_1080_24fps.mp4'
captioned_video = process_video(video_path)
print(f"Captioned video saved as {captioned_video}")
```
